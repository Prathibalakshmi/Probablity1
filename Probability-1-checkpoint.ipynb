## Coin Toss Experiment

```python
events = ['Head','Tail']   # possible events
N = [10,100,1000,10000,100000]
```

```python
head_ls = []
tail_ls = []
for n in N:
    counter = 0
    for i in range(n):
        outcome = np.random.choice(events)
        if outcome == 'Head':
            counter += 1
    head_ls.append(counter)
    tail_ls.append(n-counter)
```

```python
P_head = [head_ls[i]/N[i] for i in range(len(N))]
P_tail = [tail_ls[i]/N[i] for i in range(len(N))]
```

```python
P_head
```

```python
plt.figure(figsize=(9,6))
plt.plot(np.log10(N),P_head,label='P(heads)')
plt.plot(np.log10(N),P_tail,label='P(tails)')
plt.xlabel('log(N)')
plt.ylabel('probablity')
plt.title('Coin Toss')
plt.legend()
plt.show()
```

## Throwing a Dice

```python
event_d = [1,2,3,4,5,6]       # Possible no of events
N = [ 10,100,1000,10000,100000] # no of experiments

event1 =[]
event2 =[]
event3 =[]
event4 =[]
event5 =[]
event6 =[]

for n in N:
    count1=count2=count3=count4=count5=count6 = 0 
    for i in range(n):
        outcome = np.random.choice(event_d)        
        if outcome == 1:
            count1 +=1
        elif outcome == 2:
            count2 +=1
        elif outcome == 3:
            count3 +=1
        elif outcome == 4:
            count4 +=1
        elif outcome == 5:
            count5 +=1
        else:
            count6 +=1
        
    event1.append(count1)
    event2.append(count2)
    event3.append(count3)
    event4.append(count4)
    event5.append(count5)
    event6.append(count6)
    

print('1 Count == ',event1)
print('\n2 Count == ',event2)
print('\n3 Count == ',event3)
print('\n4 Count == ',event4)
print('\n5 Count == ',event5)
print('\n6 Count == ',event6)
```

```python
P_1 = [ event1[i]/N[i] for i in range(len(N))]
P_2 = [ event2[i]/N[i] for i in range(len(N))]
P_3 = [ event3[i]/N[i] for i in range(len(N))]
P_4 = [ event4[i]/N[i] for i in range(len(N))]
P_5 = [ event5[i]/N[i] for i in range(len(N))]
P_6 = [ event6[i]/N[i] for i in range(len(N))]

print('Probability Count 1== ',P_1)
print('\nProbability Count 2== ',P_2)
print('\nProbability Count 3== ',P_3)
print('\nProbability Count 4== ',P_4)
print('\nProbability Count 5== ',P_5)
print('\nProbability Count 6== ',P_6)
```

```python
plt.figure(figsize =(10,8))
plt.plot(np.log10(N),P_1,label = '1')
plt.plot(np.log10(N),P_2,label = '2')
plt.plot(np.log10(N),P_3,label = '3')
plt.plot(np.log10(N),P_4,label = '4')
plt.plot(np.log10(N),P_5,label = '5')
plt.plot(np.log10(N),P_6,label = '6')

plt.xlabel('N')
plt.ylabel('Probability')
plt.title('Probability Chart for a Dice')
plt.legend()
plt.show()
```

## Set Theory

### Intersection of Circles having random points

```python
def get_random_vector_in_range(a,b,n):
    arr = a+(b - a)*np.random.random(n)
    return arr
```

```python
def generate_random_points_in_two_circle(r,center1,center2,n):
    theta = np.linspace(-np.pi,np.pi,n)
    x1 = center1[0] + r * np.cos(theta)
    y1 = center1[1] + r * np.sin(theta)
    xx1 = center1[0] + get_random_vector_in_range(0,r,n)*np.cos(theta) # points inside circle
    yy1 = center1[1] + get_random_vector_in_range(0,r,n)*np.sin(theta)
    
    x2 = center2[0] + r * np.cos(theta)
    y2 = center2[1] + r * np.sin(theta)
    xx2 = center2[0] + get_random_vector_in_range(0,r,n)*np.cos(theta) # points inside circle
    yy2 = center2[1] + get_random_vector_in_range(0,r,n)*np.sin(theta)

    return np.matrix([x1,y1]).T, np.matrix([xx1,yy1]).T, np.matrix([x2,y2]).T, np.matrix([xx2,yy2]).T
```

```python
center1 = [1,0]; center2=[8,0]; radius = 5; n=1000;
circle_1_border, circle_1_inside, circle_2_border, circle_2_inside =  generate_random_points_in_two_circle(radius,center1,center2,n)
```

```python
plt.figure(figsize =(9,6))
plt.plot(circle_1_border[:,0],circle_1_border[:,1],'black')
plt.plot(circle_2_border[:,0],circle_2_border[:,1],'black')
plt.scatter(np.ravel(circle_1_inside[:,0]),np.ravel(circle_1_inside[:,1]),color='r',s=10)
plt.scatter(np.ravel(circle_2_inside[:,0]),np.ravel(circle_2_inside[:,1]),color='b',s=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Scatter Plot")
plt.show()
```

```python
X = np.vstack((circle_1_inside,circle_2_inside))   ### Dataset for al the points inside the two circles
X.shape
```

```python
np.ravel(X[0])
```

```python
common = 0
for i in range(X.shape[0]):
    dist1 = np.linalg.norm(np.ravel(X[i])-np.array(center1)) ## euclidean distance from center of first circle
    dist2 = np.linalg.norm(np.ravel(X[i])-np.array(center2)) ## euclidean distance from center of second circle
    if dist1 < radius and dist2 < radius:
        common +=1
```

## Getting events for any custom PMF

```python
def custom_distribution(events, pmf, n):   ## for discrete events only
    if len(events) == len(pmf):
        x = np.random.choice(events, p=pmf, size=n)
        return x
    else: 
        print("Both lists should be of same length!")
```

```python
n = 10000
events = ['Heads','Tails']
pmf = [0.4,0.6]
```

```python
x = custom_distribution(events=events, pmf=pmf, n=n)
x.shape
```

```python
count_heads=0; count_tails=0;
for item in x:
    if item == 'Heads':
        count_heads += 1
    else:
        count_tails += 1
```

## Bayes Theorem

```python
def bayes_theorm(p_a, p_b_given_a, p_b_given_not_a):
    # P(notA)
    p_not_a = 1 - p_a
    # P(B)
    p_b = p_a*p_b_given_a + p_not_a*p_b_given_not_a
    # P(A/B)
    p_a_given_b = (p_b_given_a*p_a)/p_b
    return p_a_given_b
```

```python
p_a = 0.02      
p_b_given_a = 0.98
p_b_given_not_a = 0.005

bayes_theorm(p_a, p_b_given_a, p_b_given_not_a)
```

## Chebyshev's Ineqaulity

```python
def calc_outliers_probability(data, lam):
    mu = np.mean(data)
    sig = np.std(data)
    n = len(data)
    
    count = 0
    for i in range(n):
        if abs(data[i] - mu) >= lam*sig:
            count += 1
    return count/n
```

```python
X = np.random.normal(-4, 5, 1000)   
lam = 1.5
calc_outliers_probability(X, lam)
```
